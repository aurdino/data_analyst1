```python
from datetime import datetime
import pandas as pd 
import requests 


url = "https://arbeitnow.com/api/job-board-api"
payload = {}
headers = {}
res = requests.request("GET",url,headers = headers, data = payload)
data = res.json()
df = pd.DataFrame(data['data'])
```
